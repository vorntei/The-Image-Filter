```python
#yk ah nis jea ah trv
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
random.seed(1)

#The Scenario
img = mpimg.imread("photo2.jpg")
print(img.shape)
print(img.ndim)
print(img.dtype)
canvas = np.zeros((255, 255, 4))

#slicing 
img2 = img[140:800, 50:700]
plt.imshow(img)
plt.axis('off')
plt.show()

# Red channel isolation (indexing)
red_channel = img2.copy()
red_channel[:,:,1] = 0
red_channel[:,:,2] = 0
#red_channel [:,:,3]=1
# print(red_channel)
plt.imshow(red_channel)
plt.axis('off')
plt.show()

# Green screen manipulation(masking) 
result = img2.copy() 
masking_screen = result[:, :, 1]> 150
random_cropped_green_img = np.random.randint(0,256, size = (len(result[masking_screen])))
result[masking_screen, 1] = random_cropped_green_img
plt.imshow(result)
plt.axis("off")
plt.show()

#BRIGHTNESS (BROADCASTING) -
bright = np.clip(img2.astype(np.int32) + 50, 0, 255).astype(np.uint8)
plt.imshow(bright)
plt.axis('off')
plt.show()

#GAMMA / VINTAGE FILTER (UFUNCS) 
normalized = img2.astype(np.float32) / 255.0
vintage = np.sqrt(normalized)
vintage = (vintage * 255).astype(np.uint8)
plt.imshow(vintage)
plt.axis('off')
plt.show()

#BLANK CANVAS
plt.imshow(canvas)
plt.axis('off')
plt.show()
```
