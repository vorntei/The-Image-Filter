```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
random.seed(1)

#The Scenario
img = mpimg.imread("photo2.jpg")
# print(img.shape)
# print(img.ndim)
# print(img.dtype)
canvas = np.zeros((350, 350, 4))

#Indexing, Slicing & Masking: 
# print(img[200, 300]) # indexing
# print(img[100:500, 200:600]) # slicing, start:end, we take at the inex end - 1
#slicing 
img2 = img[140:800, 50:700]
# plt.imshow(img)
# plt.axis('off')
# plt.show()
# Red channel isolation (indexing)
red_channel = img2.copy()
red_channel[:,:,1] = 0
red_channel[:,:,2] = 0
#red_channel [:,:,3]=1
#print(red_channel)
# plt.imshow(red_channel)
# plt.axis('off')
# plt.show()
# Green screen manipulation(masking) 
result = img2.copy() 
masking_screen = result[:, :, 1]> 150
random_cropped_green_img = np.random.randint(0,256, size = (len(result[masking_screen])))
result[masking_screen, 1] = random_cropped_green_img
plt.imshow(result)
plt.axis("off")
plt.show()
#Vectorization, Broadcasting & Ufuncs
if img2.max() <= 1.0:
    img2 = (img2 * 255).astype(np.uint8)
bright_img = img2 + 50
bright_img = np.clip(bright_img, 0, 255)
bright_img = bright_img.astype(np.uint8)
plt.imshow(bright_img)
plt.axis("off")
plt.show()
gamma = 0.5  # <1 = brighter, >1 = darker

img_norm = img2 / 255.0
gamma_img = np.power(img_norm, gamma)
gamma_img = (gamma_img * 255).astype(np.uint8)

plt.imshow(gamma_img)
plt.axis("off")
plt.show()
img_norm = img2 / 255.0
vintage = np.sqrt(img_norm)
vintage = (vintage * 255).astype(np.uint8)

plt.imshow(vintage)
plt.axis("off")
plt.show()
```
